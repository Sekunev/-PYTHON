```python
def faktöryelim(x):
  sonuç = 1
  for i in range(1, x+1):
    sonuç *= i
  if x == 0 or x == 1:
    return 1
  else :
    return sonuç
```

```python
def factor(x):
  result = 1
  for i in range(1, x+1):
    result *= i
  return print(result)
factor(5)
```

```python
def factori(x):
    result = 1
    for i in range(x):
        result *= (i+1)
    return print(result)
factori(99)
```

```python
from timeit import timeit
```

```python
timeit() # Bu modül küçük kod parçalarının çalışma sürelerini ölçmeyi sağlar.
```

```python
def for_loop():  # for loop ile list comp. arasındaki hız farkını görelim
    result = []
    
    for i in range(1000000):
        result.append(i)
    return result
```

```python
def list_comp():
    return [i for i in range(1000000)]
```

```python
size = 100
time_for =timeit(for_loop, number=size)
time_list_comp = timeit(list_comp, number = size)
```

```python
print(time_for)
print(time_list_comp)
```

```python
print(f"List_Comp. is {round(time_for / time_list_comp, 2)}) times faster than for loops")
```

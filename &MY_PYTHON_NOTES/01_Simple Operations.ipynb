
# **The Strength of Strings in Python**  [number] >> index'leme [start:stop:step] >> slice'lama

```python
fruit = 'Orange'  # String elemanlarına(karakterlerine) ayrılabilir. 
print('Kelime                    : ' , fruit)
print('Birinci Harfi             : ' , fruit[0])     # köşeli parentezle. İndex leme.Birinci Harfi
print('İkinci Harfi              : ' , fruit[1])
print(                                 fruit[:])     # hepsini yazar
print('3 le 5 arası Harfi        : ' , fruit[2:5])   # Bunun adı slayslama. 3 le 5 arası Harfi (!Unutmayın, bir dizenin numaralandırılması sıfırdan başlar)
print('3 ten sonraki Harfleri    : ' , fruit[2:])    # 3 den sonraki Harfleri 
print('6 dan önceki harfleri     : ' , fruit[:5])    # 5 ten önceki harfleri
print('2şer atlama               : ' , fruit[::2])   # 2 şer atlayarak harfler
print('Son üç harfi              : ' , fruit[-3:])   # Son üç harfi
print('Tersten yazılışı          : ' , fruit[::-1])  # Tersten yazılış
print('Fruit ', len(fruit), ' karakterlidir.')       # len() kaç karakterli olduğunu bulmaya yarar.# integer ve float değerlerde len fonksiyonu çalışmayacaktır.
print(fruit[1:-2])                                   # bu şekilde farklı kullanımları da var. İnceleyiniz...
print(fruit[2:5:-1])                                 # soldan sağa okuma yapmamız gerekiyor, ama -1 olduğu için boş sonuç verecek
print(fruit[-4])
fruit[2]
fruit[0:3]
```

```python
sentence = "I live and work in Virginia"

print(sentence.upper())     # str.upper() dizenin tüm karakterlerini büyük harfe dönüştürür.

print(sentence.lower())     #str.lower()dizenin tüm karakterlerini küçük harfe dönüştürür.

print(sentence.swapcase())  #str.swapcase()büyük harfi küçük harfe çevirir ve bunun tersi de geçerlidir.

print(sentence.capitalize())  # str.capitalize()dizenin ilk karakterini büyük harfe ve geri kalanını küçük harfe değiştirir.

print(sentence.title())  # str.title()her kelimenin ilk karakterini büyük harfe dönüştürür.

print(sentence.count("a"))  # str.count() cümledeki tırnak içerisindeki eleman sayısını sayar ve yazdırır.

print(max(sentence))  # max() en büyük değeri yazar.

print(sentence.find("a"))  # find("") karakterin stringde hangi indexte olduğunu yazdırır.

print("Virginia" in sentence)  # Sentence içerisinde Virginia varmı? sentence yerine " " içerisinde str de yazılabilirdi.

changed_sentence = sentence.replace("i", "+")  #str.replace(old, new[, count]) eskinin tüm oluşumlarını yeniyle değiştirir.
print(changed_sentence)  

print(sentence)  # note that, source text is unchanged

print(sentence.find("a"))  # find("") karakterin stringde hangi indexte olduğunu yazdırır.
print(sentence.index('a'))
```

```python
cümle = "   Bayramın ilk günü   nü"  
print(cümle.strip())              #str.strip(): her iki taraftaki tüm boşlukları (veya belirtilen karakterleri) kaldırır.
print(cümle.rstrip('mü'))             #str.rstrip() : removes spaces (or specified characters) from the right side
print(cümle.lstrip())             #str.lstrip(): sol taraftaki boşlukları (veya belirtilen karakterleri) kaldırır.
```

```python
Bayram = "Ramazan"
print(Bayram.strip("Rn"))  # strip fonksiyonu başta ve sonda istenen karakterleri kaldırdı. Tüm kombinasyonlar, yani r, n, rn seçeneklerinin hepsi.
print(Bayram.strip("Rna"))
print(Bayram.rstrip("anaz"))  # rstrip sağdaki karakteri kaldırdı
print(Bayram.lstrip("rR")) # lstrip soldaki karakteri kaldırdı.
print(Bayram.strip("naR"))
```

```python
x = "bu akşam python dersi var"
b = x.split(" ")                # split metodu uygulandığı stringin " " içerisinde belirtilen karakterlerden parçalara böldü.
print(b)
c = " ".join(b)                 # .join()metodu bir ittrable (str) ye uygulandı "" içerisinde belirtilen karakterlerden parçaları birleştirerek str oluşturdu.
print(c)
print(type(b))                  # split metodu uygulandığı stringi list türüne çevirdi.
print(type(c))
```

```python
string = "Her gün, her hafta, bilgime yeni bilgiler katıyorum, bu beni çok motive ediyor."
a = string.split()
print(a)
b = chr(37).join(a)  # chr(37)= %
print(b)
c = b.split(',%')
print(c)
d = chr(35).join(c)  # chr(35)= #
print(d, type(d))
```

```python
m, n = map(int, input('arası boşluklu 2 sayı giriniz: ').split())

for i in range(1, m, 2):
  print(str('.|.'*i).center(n, '-'))
print('WELCOME'.center(n, '-'))
for i in range(m-2, 0, -2):
  print(str('.|.'*i).center(n, '-'))
```

```python
def solve(s):
  for i in s.split():
    s = s.replace(i, i.capitalize())
  return s
solve('ali demir')
```

```python
import textwrap

def wrap(string, max_width):
    # Text Wrap in Python - HackerRank Solution START
    return textwrap.fill(string, max_width)
    # Text Wrap in Python - HackerRank Solution END
print(wrap('ABCDEFGHIJKLIMNOQRSTUVWXYZ', 4))
```
